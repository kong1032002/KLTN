```python
import pandas as pd;
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/Dataset_v1.9.csv")
df.dropna(inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df["Label"] = label_encoder.fit_transform(df["Label"])
df.head()
```
